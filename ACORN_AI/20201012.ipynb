```python
#문제 
cond1 = np.array([1,0,1,1,0,0,1], dtype=bool)
cond2 = np.array([0,1,1,1,0,1,1], dtype=bool)
#두개다 참이면 0, cond1이 참이면 1, cond2가 참이면 2, 그외일때는 3으로 처리하여 출력하시오 매핑

result = []
for i in range(len(cond1)):
    if(cond1[i] and cond2[i]):
        result.append[0]
    elif cond1[i]:
        result.append[1]        
    elif cond2[i]:
        result.append[2]
    else:
        result.append[3]
         
print(result)
```

```python
#중첩된 삼항연산자
result = np.where(cond1 & cond2, 0, np.where(cond1, 1, np.where(cond2, 2, 3)))
print(result)
```

```python
import numpy as np
# 의사역행렬
A = np.array([[1,1,0],[0,1,1],[1,1,1]])#역행렬은 정방행렬에만 존재
print(A)
Ainv=np.linalg.inv(A) #역행렬
print(Ainv)
b=np.array([[2],[2],[3]]) # 해를 구한다
b
x=Ainv@b # 행렬곱
x
```

```python
#내적은 순서변경이 무관, 행렬곱은 순서변경이 안됨
x, resid, rank, s = np.linalg.lstsq(A, b) # least square (최소제곱법)
print(x)
print(resid)
print(rank)
print(s)
```

```python
"""
다음 연립방정식의 해를 푸시오(행렬곱으로)
2x+y-2z=-3
3x+z=5
x+y-z=-2
"""
A=np.array([[2,1,-2],[3,0,1],[1,1,-1]])
b=np.transpose(np.array([-3,5,-2]))

#행렬곱 파라메터 순서는 동일해야함
print(np.linalg.solve(A, b))
print(np.linalg.inv(A) @ b)
print(np.matmul(np.linalg.inv(A), b))

```

```python
# 3차원을 2차원으로
"""
-contour 등고선 그래프 ( fill : 채우다)
"""
xlist=np.linspace(-3.0, 3.0, 100)
ylist=np.linspace(-3.0, 3.0, 100) # 10000개 정점

X, Y = np.meshgrid(xlist, ylist)

Z = np.sqrt(X**2 + Y**2)
fig,ax=plt.subplots(1,1) # msk:데이터 공유 그래프
cp=ax.contourf(X, Y, Z, levels=7, cmap='Purples')
fig.colorbar(cp)
ax.set_title('Filled Contours Plot')

ax.set_ylabel('y (cm)')
plt.show()
```

```python
def fm(p):
    x, y = p
    return np.sin(x) + 0.25 * x + np.sqrt(y)
x = np.linspace(0, 10, 20)
y = np.linspace(0, 10, 20) # 400개 정점
X, Y = np.meshgrid(x,y)

Z = fm((X, Y))
x = X.flatten() # 일렬로
y = Y.flatten()

from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
fig = plt.figure(figsize=(9,6))
ax=fig.gca(projection='3d')

# 면을 생성 #r, cstride 행 열건너뛰기 # cmap=차가운데서 따뜻한 대로(파랑->레드) 
# 안티: 계단현상을 방지 (픽셀이 사각형: 해상도가 낮으면 각진현상)
surf = ax.plot_surface(X, Y ,Z, rstride=2, cstride=2, cmap=mpl.cm.coolwarm, linewidth=0.5, antialiased=True)


ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
ax.view_init(elev=40, azim=180) # 카메라 각도
fig.colorbar(surf, shrink=0.5, aspect=5)
```

```python
# 문제 3차원 그래프를 그리고 등고선 그래프로 
def f(x, y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)
fig = plt.figure()
ax=Axes3D(fig)
X=np.arange(-4,4,0.25)
Y=np.arange(-4,4,0.25)
X, Y = np.meshgrid(X, Y)
# print(np.sin(f(X, Y)))
Z = np.sin(f(X, Y))
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
plt.show()
```
